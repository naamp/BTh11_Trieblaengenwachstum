## 1. Analyse Report Data aus Metashape
Für die folgenden Unterkapitel werden Metadaten aus Metashape verwendet. Diese Metadaten bestehen aus Berechnungsparametern oder berechneten Anzahl Punkten oder Filtering Tresholds. ### 1.1 Import der CSV-Daten
Prüfung des CSV-Imports

```python
# Laden des CSV Files
csv_path = "D:\\BTh11_Asuwertung_PC\\fhnw_code\\BTh11_Trieblaengenwachstum\\06_UAV_Comparison\\60_BaseData\\CSV\\Vergleich_DJIMini3Pro_DJIPhantom4ProV2.csv"

# Alle CSVs einlesen und in einer Liste speichern
csv_file = pd.read_csv(csv_path)

# Vorschau
display(HTML(csv_file.head(2).to_html(max_cols=200, notebook=True)))
```

### 1.2 Vergleich Root Mean Square Error (RMSE) pro UAV-Modell
Beschreibung der Visualisierung!

```python
# UAV_ID als kategoriale Variable setzen
csv_file["UAV_ID"] = pd.Categorical(
    csv_file["UAV_ID"],
    categories=["DJIMini3Pro", "DJIPhantom4ProV2"],
    ordered=True
)

# Definition der Attribute: (Spaltenname, Plot-Titel, y-Achsen-Beschriftung, y-Achsen-Limit)
attributes = [
    ("Control Points XY error [mm]", "Lagefehler pro UAV", "XY Error [mm]", (0, 8)),
    ("Control Points Z error [mm]", "Höhenfehler pro UAV", "Z Error [mm]", (0, 8)),
    ("Control Points Total RMSE [mm]", "3D-Lagefehler pro UAV", "Total RMSE [mm]", (0, 8)),
    ("Control Points Total RMSE [pix]", "Pixelfehler pro UAV", "Total RMSE [pix]", (0, 3)),
]

# Farbzuweisung für UAV_ID
aufnahme_colors = {
    "DJIMini3Pro": "#1f77b4",        # Blau
    "DJIPhantom4ProV2": "#2ca02c",    # Grün
}

# Erstelle Subplots: je ein Diagramm pro Attribut
fig, axes = plt.subplots(1, 4, figsize=(22, 5))

# Erstelle Balkendiagramme für jedes Attribut
for ax, (attr, title, ylabel, ylim) in zip(axes, attributes):
    # Für jede UAV-ID wird ein Balken gezeichnet
    for uav in csv_file["UAV_ID"].cat.categories:
        # Extrahiere den Wert für das jeweilige Attribut
        value = csv_file[csv_file["UAV_ID"] == uav][attr].values[0]
        ax.bar(uav, value, color=aufnahme_colors[uav], width=0.5)
    ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.set_xlabel("UAV Modell")
    ax.set_ylim(ylim)
    ax.grid(True, axis="y")

# Gesamttitel für alle Plots
fig.suptitle("Vergleich Root Mean Square Error (RMSE) pro UAV", fontsize=18, fontweight="bold")

plt.tight_layout()
plt.show()
```

### 1.3 Vergleich Root Mean Square Error (RMSE) pro Tie Point Filtering (TPF) Kategorie - Detailansicht Total Error

```python
# UAV_ID als kategoriale Variable setzen
csv_file["UAV_ID"] = pd.Categorical(
    csv_file["UAV_ID"],
    categories=["DJIMini3Pro", "DJIPhantom4ProV2"],
    ordered=True
)

# Definition der Attribute: (Spaltenname, Plot-Titel, y-Achsen-Beschriftung, y-Achsen-Limit)
attributes = [
    ("Control Points XY error [mm]", "Lagefehler pro UAV", "XY Error [mm]", (0, 8)),
    ("Control Points Z error [mm]", "Höhenfehler pro UAV", "Z Error [mm]", (0, 6)),
    ("Control Points Total RMSE [mm]", "3D-Lagefehler pro UAV", "Total RMSE [mm]", (0, 8)),
    ("Control Points Total RMSE [pix]", "Pixelfehler pro UAV", "Total RMSE [pix]", (0, 3)),
]

# Farbzuweisung für UAV_ID
aufnahme_colors = {
    "DJIMini3Pro": "#1f77b4",        # Blau
    "DJIPhantom4ProV2": "#2ca02c",    # Grün
}

# Erstelle für jedes Attribut einen eigenen Detailplot
for attr, title, ylabel, ylim in attributes:
    fig, ax = plt.subplots(figsize=(6, 5))
    
    # Für jede UAV-ID wird ein Balken gezeichnet
    for uav in csv_file["UAV_ID"].cat.categories:
        # Extrahiere den Wert für das jeweilige Attribut
        value = csv_file[csv_file["UAV_ID"] == uav][attr].values[0]
        ax.bar(uav, value, color=aufnahme_colors[uav], width=0.5)
    
    ax.set_title(title, fontsize=20, fontweight="bold")
    ax.set_ylabel(ylabel)
    ax.set_xlabel("UAV Modell", fontweight= "bold")
    ax.set_ylim(ylim)
    ax.grid(True, axis="y")
    
    plt.tight_layout()
    plt.show()
```
